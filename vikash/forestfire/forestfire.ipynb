```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("forestfires.csv")

# Histogram
plt.hist(df['temp'], bins=10)
plt.title("Temperature Distribution")
plt.show()

# Scatter plot
sns.scatterplot(x='temp', y='area', data=df)
plt.title("Temp vs Burned Area")
plt.show()

# Box plot
sns.boxplot(x='month', y='temp', data=df)
plt.title("Month vs Temperature")
plt.show()
```
