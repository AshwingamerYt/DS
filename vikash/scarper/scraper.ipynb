```python
import requests
from bs4 import BeautifulSoup
import random

# Step 1: Website
url = "https://quotes.toscrape.com/"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

quotes = soup.find_all("div", class_="quote")

# Step 2: Loop through data
for q in quotes:
    
    # Customer Name (Author)
    name = q.find("small", class_="author").text
    
    # Review (Quote)
    review = q.find("span", class_="text").text
    
    # Tags
    tags = [tag.text for tag in q.find_all("a", class_="tag")]
    
    # Rating (Generate realistic rating)
    rating = random.choice(["1 star", "2 star", "3 star", "4 star", "5 star"])
    
    # Review Title (short tag-based title)
    title = tags[0].capitalize() + " Thoughts" if tags else "General Review"
    
    # Step 3: Print (Amazon-style)
    print("=" * 60)
    print(f"Customer Name : {name}")
    print(f"Rating        : {rating}")
    print(f"Review Title  : {title}")
    print(f"Review        : {review}")
    print(f"Tags          : {', '.join(tags)}")
    print("=" * 60)
    print("\n")
```
